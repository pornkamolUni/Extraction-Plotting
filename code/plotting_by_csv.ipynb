```python
import pandas as pd
import plotly.graph_objects as go
import glob
import os

def display_interactive_spectrum(file_path):
    """
    Imports a CSV and displays an interactive Raman spectrum with hover-labels.
    """
    # 1. Load the data
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return

    # Check for correct column names (Generic check)
    if 'Wavenumber' not in df.columns or 'Intensity' not in df.columns:
        print(f"Skipping {file_path}: Ensure columns are 'Wavenumber' and 'Intensity'.")
        return

    # 2. Create Interactive Plot
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=df['Wavenumber'],
        y=df['Intensity'],
        mode='lines',
        line=dict(color='blue', width=1.5),
        name='Raman Intensity',
        # This creates the hover effect you requested
        hovertemplate='Wavenumber: %{x:.2f} cm⁻¹<br>Intensity: %{y:.4f}<extra></extra>'
    ))

    # 3. Formatting
    file_name = os.path.basename(file_path)
    fig.update_layout(
        title=f"Interactive Spectrum: {file_name}",
        xaxis_title="Wavenumber (cm⁻¹)",
        yaxis_title="Intensity (normalized)",
        template="plotly_white",
        hovermode="x unified",  # Shows the value as you move the mouse along the x-axis
        # Standard Raman descending scale
        #xaxis=dict(autorange="reversed")
    )

    fig.show()

# --- Automatically find and plot all CSVs ---
csv_files = glob.glob("/content/curcumin_data.csv")

for file in csv_files:
    print(f"Displaying interactive plot for: {file}")
    display_interactive_spectrum(file)
```
